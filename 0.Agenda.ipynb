## Setup


Some of the exercises on this workshop run some quantum programs on the Azure cloud, either on a simulator or on an actual quantum device.

All UW students can get a free Azure account with no credit card required by signing up at:
https://azure.microsoft.com/en-us/free/students/

Once you have your subscription, you'll need to create an Azure Quantum Workspace. Instructions can be found [here](https://docs.microsoft.com/en-us/azure/quantum/how-to-create-workspace) (note that the experience has been simplified so some of the steps are not needed anymore).

Finally, use the Notebooks section on the Azure Portal to run these notebooks. This will make it easier to connect to your online resources and to get all the notebook dependencies, like the Q# kernel.

1. Open the Azure Portal
2. Find your Azure Quantum Workspace
3. Find the Notebooks section on the left toolbar.
4. Under "My Notebooks" click the "..."
5. Upload all the notebooks in this repo.
